### **Feature Engineering**

```python
import pandas as pd

melb_data = pd.read_csv("data/melb_data_ps.csv", sep=",")
melb_data.head()
```

```python
# Старайтесь всегда оставлять переменную с первоначальным DataFrame неизменной, 
# создавайте копию исходной таблицы и совершайте преобразования на ней. 
# Это оградит вас от ошибок, которые можно совершить при подготовке данных. 

melb_df = melb_data.copy()
melb_df.head()
```

```python
melb_df = melb_data.copy()
melb_df = melb_df.drop(['index', 'Coordinates'], axis=1) #можно также доавить inplace=False, что создаст новую таблицу
melb_df.head()
```

### **Математические операции со столбцами**

```python
total_rooms = melb_df["Rooms"] + melb_df["Bedroom"] + melb_df["Bathroom"]
display(total_rooms)
```

```python
melb_df["MeanRoomSquare"] = melb_df["BuildingArea"] / total_rooms
display(melb_df["MeanRoomSquare"])
```

```python
diff_area = melb_df["BuildingArea"] - melb_df["Landsize"]
sum_area = melb_df["BuildingArea"] + melb_df["Landsize"]
melb_df["AreaRatio"] = diff_area / sum_area
display(melb_df["AreaRatio"])
```

```python
# Напишите функцию delete_columns(df, col=[]), которая удаляет столбцы из DataFrame 
# и возвращает новую таблицу. Если одного из указанных столбцов не существует в таблице, то функция должна возвращать None.

customer_df = pd.DataFrame({
            'number': [0, 1, 2, 3, 4],
            'cust_id': [128, 1201, 9832, 4392, 7472],
            'cust_age': [13, 21, 19, 21, 60],
            'cust_sale': [0, 0, 0.2, 0.15, 0.3],
            'cust_year_birth': [2008, 2000, 2002, 2000, 1961],
            'cust_order': [1400, 14142, 900, 1240, 8430]
        })


# Вариант ALD
def delete_columns(df, col=[]):
    for i in col:
        if i not in df:
            return None
        else: continue
    return df.drop(col, axis=1, inplace=False)
    

print(delete_columns(customer_df, col=["number", "sasa"]))
```

```python
# Для каждой страны рассчитайте плотность населения (количество человек на квадратный километр).
# Затем по полученным данным рассчитайте среднее по плотностям населения в указанных странах. Ответ округлите до сотых.
# Плотность населения рассчитывается как количество человек, проживающих на территории отдельной страны, 
# делённое на площадь этой страны. Обратите внимание, что население в таблице представлено в миллионах.

import pandas as pd

countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'square': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})

countries_df2 = countries_df.copy()
countries_df2["PeoplePerSquare"] = countries_df2["population"] / countries_df2["square"] * 1000000
display(countries_df2)
display(countries_df2["PeoplePerSquare"].mean())
```

### **Признаки даты и времени**

```python
import pandas as pd

melb_data = pd.read_csv("data/melb_data_ps.csv", sep=",")
melb_df2 = melb_data.copy()

display(melb_df2["Date"])
```

```python
melb_df2 = melb_data.copy()
melb_df2["Date"] = pd.to_datetime(melb_df2["Date"], dayfirst=True)

display(melb_df2["Date"])
```

```python
years_sold = melb_df2["Date"].dt.year
print(years_sold)
print("Min year sold", years_sold.min())
print("Max year sold", years_sold.max())
print("Mode year sold", years_sold.mode()[0])
```

```python
melb_df2["MonthSale"] = melb_df2["Date"].dt.month
melb_df2["MonthSale"].value_counts(normalize=True)
```

```python
delta_days = melb_df2["Date"] - pd.to_datetime("2016-01-01")
display(delta_days)
display(delta_days.dt.days)
```

```python
melb_df2["AgeBuilding"] = melb_df2["Date"].dt.year - melb_df2["YearBuilt"]
display(melb_df2["AgeBuilding"])
```

```python
melb_df2 = melb_df2.drop("YearBuilt", axis=1)
```

```python
melb_df2["WeekdaySale"] = melb_df2["Date"].dt.day_of_week

weekend_count = melb_df2[melb_df2["WeekdaySale"] == 5].shape[0] + melb_df2[melb_df2["WeekdaySale"] == 6].shape[0]
print(weekend_count)
```

### **Задача с НЛО**

```python
import pandas as pd

ufo = pd.read_csv("data/ufo.csv", sep=",")


# В каком году отмечается наибольшее количество случаев наблюдения НЛО в США?
ufo["Time"] = pd.to_datetime(ufo["Time"], dayfirst=False)
ufo["Year"] = ufo["Time"].dt.year
ufo["Year"].value_counts()
```

```python
# Найдите средний интервал времени (в днях) между двумя последовательными случаями наблюдения НЛО в штате Невада (NV).

ufo["Date"] = ufo["Time"].dt.date
ufo["DaysDiff"] = ufo["Date"].diff()
```
