# Продвинутые методы работы с данными в Pandas

```python
import pandas as pd

melb_df = pd.read_csv("data/melb_data_fe.csv")
melb_df.head()
```

```python
melb_df.info()
```

```python
melb_df["Date"] = pd.to_datetime(melb_df["Date"])
melb_df_quarter = melb_df["Date"].dt.quarter
melb_df_quarter.value_counts()
```

```python
exception_col = ["Date", "Rooms", "Bedroom", "Bathroom", "Car"]
max_unique = 150

for col in melb_df:
    if melb_df[col].nunique() < max_unique and col not in exception_col:
        melb_df[col] = melb_df[col].astype("category")
        
melb_df.info()
```

```python
melb_df.sort_values(by="Price", ascending=False).head(10)
```

```python
melb_df.sort_values(by="Date", ascending=False)
```

```python
melb_df.sort_values(by=["Distance", "Price"]).loc[::10, ["Distance", "Price"]]
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
sorted = melb_df.copy()

sorted.sort_values(
    by=["AreaRatio"],
    ascending=False,
    ignore_index=True,
    inplace=True
)
sorted.loc[1558]
```

```python
melb_df.sort_index(inplace=True)
```

```python
melb_df.head()
```

```python
new_melb_df = melb_df.copy()

new_mask1 = new_melb_df["Type"] == "townhouse"
new_mask2 = new_melb_df["Rooms"] > 2

new_melb_df[new_mask1 & new_mask2].sort_values(
    by=["Rooms", "MeanRoomsSquare"],
    ascending=[True, False],
    ignore_index=True
).loc[18]
```

### **Группировка**

```python
melb_df.groupby(by='Type')["Price"].mean()
```

```python
melb_df.groupby("Regionname")["Distance"].min().sort_values(ascending=False)
```

```python
melb_df.groupby("MonthSale")["Price"].agg(["count", "mean", "max"]).sort_values(by="count", ascending=False)
```

```python
melb_df.groupby("MonthSale")["Price"].describe()
```

```python
melb_df.groupby("Regionname")["SellerG"].agg(["nunique", set])
```
