```python
import pandas as pd

covid_data = pd.read_csv("data/covid_data.csv")
display(covid_data.head())
```

```python
vac = pd.read_csv("data/country_vaccinations.csv")

# Если мы хотим оставить только определенные столбцы из таблицы, достаточно указать следующим образом
vac = vac[
    ['country', 'date', 'total_vaccinations', 
     'people_vaccinated', 'people_vaccinated_per_hundred',
     'people_fully_vaccinated', 'people_fully_vaccinated_per_hundred',
     'daily_vaccinations', 'vaccines']
]

vac.head()
```

```python
# Первой таблице данные представлены по каждому штату/провинции, а во второй по странам.
# Необходимо преобразовать первую таблицу и представить данные по странам.

covid_data = covid_data.groupby(
    ['date', 'country'], 
    as_index=False
)[['confirmed', 'deaths', 'recovered']].sum()
```

```python
display(covid_data.head())
```

```python
covid_data["date"] = pd.to_datetime(covid_data["date"])
```

```python
covid_data["active"] = covid_data["confirmed"] - covid_data["recovered"] - covid_data["deaths"]
```

```python
# Создадим признак ежедневного прироста числа заболевших, умерших и выздоровевших людей. 
# Для этого отсортируем данные по названиям стран, а затем по датам. 
# После этого произведём группировку по странам и рассчитаем разницу между «вчера и сегодня» с помощью метода diff():

covid_data = covid_data.sort_values(by=["country", "date"])
covid_data["daily_confirmed"] = covid_data.groupby("country")["confirmed"].diff()
covid_data["daily_deaths"] = covid_data.groupby("country")["deaths"].diff()
covid_data["daily_recovered"] = covid_data.groupby("country")["recovered"].diff()
```
